```python
def generate_decision_tree(df, feature_names, target_name, max_depth, min_samples_leaf): 

    from google.cloud import storage
    import matplotlib.pyplot as plt

    from matplotlib import pyplot as plt
    from sklearn import datasets
    from xgboost import XGBClassifier
    from xgboost import plot_tree
    from sklearn.model_selection import train_test_split

    # define X and y
    X = df[feature_names]
    y = df[target_name]
    
    # Fit the classifier with max_depth=3
    clf = XGBClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf, random_state=1234)

    model = clf.fit(X, y)

    _ = plt.figure(figsize=(120,80))
    plot_tree(model, feature_names=feature_names, class_names=target_name)
    
    plt.show()
    
    return _
```

```python
from google.cloud import storage
import matplotlib.pyplot as plt

def export_plot_to_gcs(figure, bucket_name, plot_path, plot_filename):
    # Create a client to interact with GCS
    storage_client = storage.Client()

    # Get the specified GCS bucket
    bucket = storage_client.get_bucket(bucket_name)

    # Define the GCS path where you want to save the plot
    gcs_path = f"{plot_path}/{plot_filename}"

    # Save the plot locally (e.g., to a file named 'plot.png')
    figure.savefig("plot.png")

    # Upload the plot to the GCS bucket
    blob = bucket.blob(gcs_path)
    blob.upload_from_filename("plot.png")

    # Optionally, make the plot public (readable by anyone)
    blob.make_public()

    # Delete the local plot file (if desired)
    os.remove("plot.png")

    return f"Plot exported to GCS: gs://{bucket_name}/{gcs_path}"
```
